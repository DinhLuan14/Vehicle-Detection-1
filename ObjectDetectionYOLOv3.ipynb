#### Chỉnh cấu hình file

1. Trong file yolov3_training.cfg:
*   batch=64
*   subdivisions=16
*   max_batches=classes*2000
*   steps = 80%classes, 90%classes
*   width và height là số chia hết cho 32 (mặc định 418x418)
*   classes = số class - (sửa trong 3 layer yolo)
*   filters = (classes + 5)x3 - (sửa trong 3 layer [convolutional] trước 3 layer [yolo])

2. Tạo file obj.names trong thư mục data - mỗi dòng là tên class

3. Tạo file obj.data trong thư mục data 
```
    classes = 14
    train  = data/train.txt
    valid  = data/test.txt
    names = data/obj.names
    backup = backup/
```
4. Tạo 2 file train.txt và obj.names (trong phần dataset)
5. Tải pre-train weight: https://pjreddie.com/media/files/darknet53.conv.74
