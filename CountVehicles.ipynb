```
# Count object in file .xml
lst = ['+','2w','2 ','02','ethnicity','religion','21','111','botleft','name','26','21']
def CountVehicle(file_xml):
  try :
    print(file_xml)
    tree = ET.parse(file_xml)
    root = tree.getroot() 
    for object in root.findall('object'):
      name = object.find('name').text
      count[name]=count[name]+1
  except:
    pass
```
